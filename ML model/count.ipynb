```python

from ultralytics import YOLO
import cv2
import os
```

```python
# YOLOv8 pretrained model
model = YOLO('yolov8n.pt')  # 'yolov8s.pt' for more accuracy
```

```python
# 3. Defining class names for vehicles and cyclists
vehicle_classes = ['car', 'bus', 'truck']
cyclist_classes = ['bicycle', 'motorcycle']
```

```python
# 4. Inference on a single image
image_path = ''  # image path
results = model(image_path)

vehicle_count = 0
cyclist_count = 0

for r in results:
    for c in r.boxes.cls:
        class_name = model.names[int(c)]
        if class_name in vehicle_classes:
            vehicle_count += 1
        elif class_name in cyclist_classes:
            cyclist_count += 1

print(f'Image: Vehicles: {vehicle_count}, Cyclists: {cyclist_count}')

# Showing the image with detections
results[0].show()
```
